## Q1. What is the difference between "Love" and "love" in NLP?

In NLP, text is case-sensitive by default, which means "Love" and "love" are treated as completely different words. This can mess things up because it increases the number of unique words in the dataset and makes the model less efficient. To avoid this, we usually convert everything to lowercase during preprocessing so both "Love" and "love" are treated the same.

## Q2. What happens if stopwords are not removed?


If we don't remove stopwords, the dataset ends up full of common words like "is", "the", and "and" that don't really add much meaning. This makes the data more bulky and noisy. Due to which, models take more time to process the data and may not perform as well since they're focusing on useless words instead of meaningful ones.

## Q3. Two real-world scenarios where removing stopwords can be harmful:


In sentiment analysis, removing words like "not" can completely change the meaning. For example, "not good" would become just "good". This is obviously wrong!

In question-answer systems, removing words such as who, what or where takes away the main intent of the question making it harder for a model to understand what's being asked.

## Q4. Difference between Stemming and Lemmatization:


Stemming: A rule-based approach where we remove prefixes or suffixes to get to some base form of a word without considering its context.

Examples: "running" → "run", "studies" → "studi" (may not be a real word)
Lemmatization: An advanced process that takes into account the context of a word before converting it into its actual dictionary base form so that the output is always a proper word.

Examples: "running" → "run", "studies" → "study", "better" → "good"

## Task 2: Build Advanced Preprocessing Function

```
import re

def preprocess_text(text):

    # handling invalid input
    if not text or not isinstance(text, str):
        return [], ""

    # converting text to lowercase
    text = text.lower()

    # removing URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # removing email patterns
    text = re.sub(r"\S+@\S+", "", text)

    # removing numbers
    text = re.sub(r"\d+", "", text)

    # removing special characters and emojis
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # handling repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # removing extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # spliting into tokens
    tokens = text.split()

    # removing very short words but keeps 'no' and 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # join tokens back into a sentence
    cleaned_text = " ".join(tokens)

    return tokens, cleaned_text
```

## Task 5: Frequency Analysis

```
from collections import Counter

all_tokens = []

# combining all tokens from all sentences
for _, tokens, _ in results:
    all_tokens.extend(tokens)

freq = Counter(all_tokens)

print("Top 10 most frequent words:")
print(freq.most_common(10))

print("\nTop 5 least frequent words:")
print(freq.most_common()[-5:])

Top 10 most frequent words:
[('this', 4), ('now', 3), ('get', 1), ('free', 1), ('access', 1), ('absolutely', 1), ('looved', 1), ('product', 1), ('worst', 1), ('service', 1)]

Top 5 least frequent words:
[('limited', 1), ('offer', 1), ('not', 1), ('happy', 1), ('with', 1)]
```

## Task 6: Build Full Pipeline

```
def full_pipeline(text_list):

    #preprocessing on list of sentences

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, clean = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
